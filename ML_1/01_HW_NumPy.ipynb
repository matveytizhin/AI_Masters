Реализуйте функцию, которая считает проекцию одного вектора $a \in \mathbb{R}^n$ на другой вектор $b \in \mathbb{R}^n$.

````python
def project(a, b):
    cos_a_b = (a @ b) / np.linalg.norm(b)**2
    return b * cos_a_b
````

Напишите функцию, создающую матрицу размера $n \times n$ с единицами на главной и побочной диагоналях.

```
array([[1, 0, 0, 0, 1],
       [0, 1, 0, 1, 0],
       [0, 0, 1, 0, 0],
       [0, 1, 0, 1, 0],
       [1, 0, 0, 0, 1]])
```

````python
def double_eye(n: int, dtype: np.dtype = np.int64):
    cross_matrix = np.eye(n, n) + np.fliplr(np.eye(n, n))
    if n % 2 == 1:
        cross_matrix[n // 2, n // 2] = 1
    return cross_matrix.astype(int)
````

````python
def get_nonzero_diag_product(a):
    diag_a_without_0 = np.diag(a)[np.diag(a) != 0]
    if diag_a_without_0.shape[0] != 0:
        return np.prod(diag_a_without_0)
    return None
````

### Задача 2.2 (1 балл)

Напишите функцию `replace_nan_to_means(X)`, принимающую матрицу $X$. Функция должна вернуть копию матрицы $X$, в которой все значения `nan` в каждом столбце заменены на среднее арифметическое остальных элементов столбца. В случае столбца из одних `nan` необходимо заменить все элементы столбца на нули. Исходная матрица `X` должна остаться неизменной.

````python
def replace_nan_to_means(X):
    column_means = np.nanmean(X, axis=0)
    column_means[np.isnan(column_means)] = 0 #Если столбец только из nan
    return np.where(np.isnan(X), column_means, X)
````

````python
assert np.isclose(
    replace_nan_to_means(np.array([[0, 1, 2], [np.nan, 1, np.nan], [5, 6, 7]])),
    np.array([[0, 1, 2], [2.5, 1, 4.5], [5, 6, 7]])).all()
assert np.isclose(
    replace_nan_to_means(
        np.array(
            [[ 6.,  1.,  5.],
             [ 6.,  8.,  7.],
             [ 5.,  8.,  7.],
             [np.nan,  2.,  2.],
             [np.nan,  3.,  1.]])
    ),
    np.array(
        [[6.     , 1.     , 5.     ],
        [6.     , 8.     , 7.     ],
        [5.     , 8.     , 7.     ],
        [5.66667, 2.     , 2.     ],
        [5.66667, 3.     , 1.     ]]
    )
).all()
````

````python
def get_max_after_zero(x):
    mask_before_zero = np.zeros_like(x, dtype=bool)
    mask_before_zero[1:] = (x == 0)[:-1]
    x_for_max = x[mask_before_zero]
    if x_for_max.shape[0] != 0:
        return np.max(x_for_max)
    return None
x = np.array([1, 2, 9, 8, 0, 5])
get_max_after_zero(x)
````

### Задача 2.4 (1 балл)

Напишите функцию `encode_rle(x)`, реализующую кодирование длин серий (Run-length encoding). По входному вектору x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй — сколько раз их нужно повторить.

````python
def encode_rle(x):
    changes = x[1:] != x[:-1]
    changes_indices = np.concatenate([[0], np.where(changes)[0] + 1])
    values = x[changes_indices]
    counts = np.diff(np.concatenate([changes_indices, [len(x)]]))
    return values, counts
````

````python
import numpy as np

def encode_rle(x):
    changes = x[1:] != x[:-1]
    change_indices = np.concatenate([[0], np.where(changes)[0] + 1])
    values = x[change_indices]
    counts = np.diff(np.concatenate([change_indices, [len(x)]]))
    return (values, counts)

# Проверка с примерами
x = np.array([5, 5, 5, 2, 2, 2, 4, 4])
print(encode_rle(x))

x_single = np.array([0])
print(encode_rle(x_single))

x_mixed = np.array([1, 1, 1, 0, 0, 1, 1, 0, 1])
print(encode_rle(x_mixed))
````

````python
assert encode_rle(np.array([0])) == (np.array([0]), np.array([1]))
values, count = encode_rle(np.array([5, 5, 5, 2, 2, 2, 4, 4]))
assert (values == np.array([5,2,4])).all()
assert (count == np.array([3,3,2])).all()
values, count = encode_rle(np.array([1, 1, 1, 0, 0, 1, 1, 0, 1]))
assert (values == np.array([1, 0, 1, 0, 1])).all()
assert (count == np.array([3, 2, 2, 1, 1])).all()
````

### Задача 2.7 (1.5 балла)

Охотники за сокровищами отправились за очередным кладом в необычное место, которое называется "Поле чудес". "Поле чудес" имеет размер $H$ на $W$ метров. На каждом квадратном метре "Поля чудес" закопан сундук с определённым количеством камней. Скоро "Поле чудес" должен посетить лепрекон. Под влиянием лепрекона камни в сундуках превращаются в золотые монеты. Лепрекон появляется в определённой клетке поля и действует на прямоугольную область размером $h$ на $w$, верхний левый край этой области - точка, где стоит лепрекон. Охотники за сокровищами не знают, когда точно появятся лепрекон, но хотят получить как можно больше золота. Известно вероятностное распределение на появление лепрекона в каждый момент времени в каждой клетке поля. Необходимо для каждой клетки "Поля чудес" посчитать математическое ожидание награды, которое будет получено от этой клетки. </br>
Необходимо написать функцию `calc_expectations(h, w, X, Q)`, где:
* $h$, $w$ - размеры области влияния лепрекона;
* $X$ - матрица целых чисел размера $H$ на $W$; $X[i, j]$ - количество камней в $i$, $j$ клетке;
* $Q$ - матрица вещественных чисел размера $H$ на $W$, задающая вероятностное распределение; $Q[i, j]$ - вероятность появления лепрекона в точке $i$, $j$.
Функция возвращает матрицу $E$ размера $H$ на $W$; $E[i, j]$ – математическое ожидание награды в $i$, $j$ клетке.

````python
import numpy as np

def calc_expectations(h, w, X, Q):
    H, W = Q.shape
    P = np.zeros((H+1, W+1))
    P[1:, 1:] = np.cumsum(np.cumsum(Q, axis=0), axis=1)

    def sum_rect(r1, c1, r2, c2):
        return (P[r2+1, c2+1] - P[r1, c2+1] - P[r2+1, c1] + P[r1, c1])

    r1 = np.maximum(0, np.arange(H) - h + 1)
    c1 = np.maximum(0, np.arange(W) - w + 1)

    R1, C1 = np.meshgrid(r1, c1, indexing='ij')
    R2 = np.arange(H)[:, None]
    C2 = np.arange(W)[None, :]

    S = sum_rect(R1, C1, R2, C2)

    E = X * S
    return E
````

### Задача 2.8 (1.5 балла)

Выберите одну любую задачу из части 2. Реализуйте ее, не используя NumPy (можно использовать циклы `for`/`while` и т.п.). Можете сделать несколько новых реализаций. Сравните время работы каждой реализации на одинаковом наборе данных. Опишите свои наблюдения. Пример сравнения времени работы можете найти в семинаре №1.

Вывод наблюдений **нужно** сделать либо в таблице, либо на графике (например, с помощью библиотеки [matplotlib](https://matplotlib.org/), также подумайте сколько экспериментов нужно провести, чтобы сделать вывод (_спойлер_: не 1).

````python
import numpy as np
import time
import matplotlib.pyplot as plt

def get_nonzero_diag_product_numpy(a):
    diag_a_without_0 = np.diag(a)[np.diag(a) != 0]
    if diag_a_without_0.size != 0:
        return np.prod(diag_a_without_0)
    return None

def get_nonzero_diag_product_python(a):
    product = 1.0
    found_nonzero = False
    n = min(a.shape[0], a.shape[1])
    for i in range(n):
        element = a[i, i]
        if element != 0:
            product *= element
            found_nonzero = True
    if found_nonzero:
        return product
    return None

sizes = [5, 50, 500, 5000]

numpy_times = []
python_times = []

for size in sizes:
    matrix = np.random.randint(0, 1, size=(size, size))
    start_time = time.time()
    get_nonzero_diag_product_numpy(matrix)
    end_time = time.time()
    numpy_times.append(end_time - start_time)

    start_time = time.time()
    get_nonzero_diag_product_python(matrix)
    end_time = time.time()
    python_times.append(end_time - start_time)

plt.figure(figsize=(10, 6))
plt.plot(sizes, numpy_times, label='NumPy')
plt.plot(sizes, python_times, label='Python (циклы)')
plt.xlabel('Размер матрицы (N x N)')
plt.ylabel('Время выполнения (секунды)')
plt.title('Сравнение времени выполнения NumPy vs Python (Задача 2.1)')
plt.legend()
plt.grid()
plt.show()
````

Наблюдения:
По графику видно, что с увеличением размера матрицы время выполнения реализации с циклами растет значительно быстрее, чем у NumPy-реализации.
